```python
import numpy as np
import torch as tc
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from torchvision import datasets
!pip install umap-learn
import umap as u

root = "/content"
download = False
dataset = datasets.MNIST(root, download=download)


image, current_class = dataset[0]
image = np.array(image, dtype=np.float32)

print(f"Image dtype: {image.dtype}")
print(f"Image shape: {image.shape}")
print(f"Image size: {image.size}")

plt.figure()
plt.imshow(image, cmap='gray')
plt.title(f"Class: {current_class}")
plt.show()


num_samples = 1000
all_features = np.zeros((num_samples, 784))
all_labels = np.zeros(num_samples)
for idx in range(num_samples):
  current_image, current_class = dataset[idx]
  all_features[idx] = np.array(current_image, dtype=np.float32).ravel()
  all_labels[idx] = current_class

print(f"Shape of all features: {all_features.shape}")

umap = u.UMAP(n_components=2)
all_features_reduced_umap = umap.fit_transform(all_features)
print(f"UMAP Shape of reduced features: {all_features_reduced_umap.shape}")

tsne = TSNE(n_components=2, perplexity=30)
all_features_reduced = tsne.fit_transform(all_features)
print(f"TSNE Shape of reduced features: {all_features_reduced.shape}")

pca = PCA(n_components=2)
all_features_reduced_pca = pca.fit_transform(all_features)
print(f"PCA Shape of reduced features: {all_features_reduced_pca.shape}")


legend = [str(item) for item in range(0, 10)]
plt.figure()
plt.plot(all_features_reduced[all_labels == 0, 0], all_features_reduced[all_labels == 0, 1], "*", color='black')
plt.plot(all_features_reduced[all_labels == 1, 0], all_features_reduced[all_labels == 1, 1], "*", color='grey')
plt.plot(all_features_reduced[all_labels == 2, 0], all_features_reduced[all_labels == 2, 1], "*", color='red')
plt.plot(all_features_reduced[all_labels == 3, 0], all_features_reduced[all_labels == 3, 1], "*", color='sienna')
plt.plot(all_features_reduced[all_labels == 4, 0], all_features_reduced[all_labels == 4, 1], "*", color='gold')
plt.plot(all_features_reduced[all_labels == 5, 0], all_features_reduced[all_labels == 5, 1], "*", color='hotpink')
plt.plot(all_features_reduced[all_labels == 6, 0], all_features_reduced[all_labels == 6, 1], "*", color='yellow')
plt.plot(all_features_reduced[all_labels == 7, 0], all_features_reduced[all_labels == 7, 1], "*", color='pink')
plt.plot(all_features_reduced[all_labels == 8, 0], all_features_reduced[all_labels == 8, 1], "*", color='lime')
plt.plot(all_features_reduced[all_labels == 9, 0], all_features_reduced[all_labels == 9, 1], "*", color='skyblue')
plt.legend(legend)
plt.show()

plt.figure()
plt.plot(all_features_reduced_pca[all_labels == 0, 0], all_features_reduced_pca[all_labels == 0, 1], "*", color='black')
plt.plot(all_features_reduced_pca[all_labels == 1, 0], all_features_reduced_pca[all_labels == 1, 1], "*", color='grey')
plt.plot(all_features_reduced_pca[all_labels == 2, 0], all_features_reduced_pca[all_labels == 2, 1], "*", color='red')
plt.plot(all_features_reduced_pca[all_labels == 3, 0], all_features_reduced_pca[all_labels == 3, 1], "*", color='sienna')
plt.plot(all_features_reduced_pca[all_labels == 4, 0], all_features_reduced_pca[all_labels == 4, 1], "*", color='gold')
plt.plot(all_features_reduced_pca[all_labels == 5, 0], all_features_reduced_pca[all_labels == 5, 1], "*", color='hotpink')
plt.plot(all_features_reduced_pca[all_labels == 6, 0], all_features_reduced_pca[all_labels == 6, 1], "*", color='yellow')
plt.plot(all_features_reduced_pca[all_labels == 7, 0], all_features_reduced_pca[all_labels == 7, 1], "*", color='pink')
plt.plot(all_features_reduced_pca[all_labels == 8, 0], all_features_reduced_pca[all_labels == 8, 1], "*", color='lime')
plt.plot(all_features_reduced_pca[all_labels == 9, 0], all_features_reduced_pca[all_labels == 9, 1], "*", color='skyblue')
plt.legend(legend)
plt.show()


plt.figure()
plt.plot(all_features_reduced_umap[all_labels == 0, 0], all_features_reduced_umap[all_labels == 0, 1], "*", color='black')
plt.plot(all_features_reduced_umap[all_labels == 1, 0], all_features_reduced_umap[all_labels == 1, 1], "*", color='grey')
plt.plot(all_features_reduced_umap[all_labels == 2, 0], all_features_reduced_umap[all_labels == 2, 1], "*", color='red')
plt.plot(all_features_reduced_umap[all_labels == 3, 0], all_features_reduced_umap[all_labels == 3, 1], "*", color='sienna')
plt.plot(all_features_reduced_umap[all_labels == 4, 0], all_features_reduced_umap[all_labels == 4, 1], "*", color='gold')
plt.plot(all_features_reduced_umap[all_labels == 5, 0], all_features_reduced_umap[all_labels == 5, 1], "*", color='hotpink')
plt.plot(all_features_reduced_umap[all_labels == 6, 0], all_features_reduced_umap[all_labels == 6, 1], "*", color='yellow')
plt.plot(all_features_reduced_umap[all_labels == 7, 0], all_features_reduced_umap[all_labels == 7, 1], "*", color='pink')
plt.plot(all_features_reduced_umap[all_labels == 8, 0], all_features_reduced_umap[all_labels == 8, 1], "*", color='lime')
plt.plot(all_features_reduced_umap[all_labels == 9, 0], all_features_reduced_umap[all_labels == 9, 1], "*", color='skyblue')
plt.legend(legend)
plt.show()




```
